## What is the true normal human body temperature? 

#### Background

The mean normal body temperature was held to be 37$^{\circ}$C or 98.6$^{\circ}$F for more than 120 years since it was first conceptualized and reported by Carl Wunderlich in a famous 1868 book. In 1992, this value was revised to 36.8$^{\circ}$C or 98.2$^{\circ}$F.

```python
import pandas as pd
import pylab
import scipy.stats as stats
import numpy as np
%matplotlib inline
```

```python
df = pd.read_csv('data/human_body_temperature.csv')
```

```python
df.plot(y='temperature',kind='hist', bins=15)
```

1) **Is the distribution of body temperatures normal?**

According to the below normal probablity plot, because all the points except for one are near the line, temperature data here are approximately normal, and the condition for the CLT is met.

stats.probplot(temp, dist='norm', plot=pylab)

```python
temp = df['temperature']
stats.probplot(temp, dist='norm', plot=pylab)
```

2) **Is the true population mean really 98.6 degrees F?**

It is only appropriate to use the one sample t test in this situation, because the population standard deviation for temperatures is not known which violates a necessary condition to use the z test. According to the below t test, due to the large test statistic and very small p-value, we have strong evidence to believe that the true mean temperature is not 98.6 degrees F.

```python
print "t-statistic: ", stats.ttest_1samp(temp,98.6)[0]
print "Two tailed p-value: ", stats.ttest_1samp(temp,98.6)[1]
```

```python
se = np.std(temp)/(len(temp))**(0.5)
sm = np.mean(temp)
print "Sample Standard error: ", se
print "Sample mean temperate is ", sm
print "Margin of error is twice the standard error, or ", 2*se
print "A 95 percent confidence interval for the true mean temperature is (",sm - 2*se,",",sm + 2*se,")"
```

3) **At what temperature should we consider someone's temperature to be abnormal?**

Approximately 95 percent of all temperatures will be within 2 standard deviations of the mean. So using the above estimates (sample mean and sample standard error) which generated the above confidence interval, an "abnormal" temperate can be considered to be one below 98.1 or above 98.3 degrees.

```python
males = df[df['gender'] == 'M']
females = df[df['gender'] == 'F']
mtemp = males['temperature']
ftemp = females['temperature']
print "Two sample t test t-statistic: " , stats.ttest_ind(mtemp, ftemp)[0]
print "Two tailed p-value: " ,stats.ttest_ind(mtemp, ftemp)[1]
```

4) **Is there a significant difference between males and females in normal temperature?**

Carrying out the two sample t test for a difference in mean temperature between males and females as above yields a p-value of 0.0239. This is a small p-value, and gives strong evidence to reject the null hypothesis that the mean temperatures for males and females are the same.
